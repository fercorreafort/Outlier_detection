```python
import pandas as pd 
import numpy as np 
import plotly.express as px
```

```python
titles = pd.read_csv(r'/Users/fercorreafort/Projects/Outlier_detection git/combined_titles.csv')
```

```python
px.bar(round(titles.select_dtypes(include=np.number).isna().sum() / titles.shape[0] *100,2), 
        x=titles.select_dtypes(include=np.number).isna().sum().index,
        y=titles.select_dtypes(include=np.number).isna().sum().values,
        labels={'x': 'Columns', 'y': 'Percentage of Missing Values'},
        title='Percentage of Missing Values in Object Columns')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
docs = titles.title.fillna('')
cv = CountVectorizer(ngram_range=(2,2), stop_words='english')
X = cv.fit_transform(docs)
print(*sorted(zip(cv.get_feature_names_out(), X.sum(axis=0).A1), key=lambda x: -x[1])[:50], sep='\n')
```
