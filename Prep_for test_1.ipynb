```python
import pandas as pd 
import numpy as np
import kagglehub
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# Download latest version
path = kagglehub.dataset_download("saichaitanyareddyai/car-price-dataset")
car_df = pd.read_csv(f"{path}/car_price_dataset.csv")
car_df.columns
```

```python
import os

# Split dataframe by brand and save each to a CSV, but do not overwrite existing files
brands = car_df['Brand'].unique()
brand_dfs_dict = {brand: car_df[car_df['Brand'] == brand].copy() for brand in brands}

folder_path = r'/Users/fercorreafort/Projects/Outlier_detection git/Tests/data_1'
for brand, df in brand_dfs_dict.items():
    filename = f'{brand}.csv'
    save_path = os.path.join(folder_path, filename)
    if os.path.exists(save_path):
        print(f'Skipping existing {save_path}')
        continue
    df.to_csv(save_path, index=False)
    print(f'Saved to {save_path}')
```
