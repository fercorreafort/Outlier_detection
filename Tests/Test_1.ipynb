```python
# Load all CSVs in the data folder into a dict keyed by filename (stem)
data_frames = {}
for csv_path in sorted(data_folder.glob('*.csv')):
    try:
        df = pd.read_csv(csv_path)
        key = csv_path.stem
        data_frames[key] = df
    except Exception as e:
        print(f'Failed to read {csv_path}: {e}')

# Show which frames were loaded
list(data_frames.keys())
```
